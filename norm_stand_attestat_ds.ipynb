```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
ds = pd.read_csv('attestat_dataset.csv')
```

```python
print(ds.columns)
print(ds.sample(5)) # or head(5)
```

```python
ds.describe()
```

```python
ds.info()
```

```python
ds.drop(columns = ['PassengerId', 'Pclass', 'Name', 'Ticket', 'Cabin', 'Embarked'], axis = 1, inplace=True)
```

```python
ds.dropna(inplace=True)
```

```python
encoders = {}
encoder = LabelEncoder()
encoder.fit(ds['Sex'])
ds['Sex'] = encoder.transform(ds['Sex'])
encoders['Sex'] = encoder
```

```python
ds.dtypes
```

```python
numerics = ['int32', 'int64', 'float64']
newdf = ds.select_dtypes(include = numerics)
```

```python
ds.describe()
```

```python
# Инициализация нормализатора
minmax_scaler = MinMaxScaler()

#Обучение MinMaxScaler
dataset_minmax = minmax_scaler.fit_transform(np.array(ds[['Survived', 'Sex', 'Age',  'SibSp', 'Parch', 'Fare']]))
```

```python
#возвращаем названия столбцов
dataset_minmax_df = pd.DataFrame(minmax_scaler.fit_transform(ds), columns = ds.columns)
dataset_minmax_df.head()
```

```python
dataset_minmax_df.describe()
```

```python
#StandardScaler
std_scaler = StandardScaler()
dataset_std = std_scaler.fit_transform(np.array(ds[['Survived', 'Sex', 'Age',  'SibSp', 'Parch', 'Fare']]))
```

```python
dataset_std[:1]
```

```python
dataset_std_df = pd.DataFrame(data = dataset_std, columns = ['Survived', 'Sex', 'Age',  'SibSp', 'Parch', 'Fare'])
```

```python
dataset_std_df.describe()
```

```python
total_sum_norm = dataset_minmax_df.sum().sum()
```

```python
total_sum_std = dataset_std_df.sum().sum()
```

```python
rounded_sum_1 = int(round(total_sum_norm))
rounded_sum_2 = int(round(total_sum_std))
```
